```python
from re import L
Q1 = df3['Close'].quantile(0.25)
Q3 = df3['Close'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df3[(df3['Close'] < lower_bound) | (df3['Close'] > upper_bound)]
outliers
```

```python
media_close = df3['Close'].mean()
for outliers in outliers:
  df3['Close'].replace(outliers,media_close,inplace=True)
  df3['Close'] = df3['Close'].replace(outliers,media_close)
df3
```

```python
total_rows = len(df3)
legs =21

df_train = df3.iloc[:total_rows - legs]
df_test = df3.iloc[total_rows - legs:]

print(len(df_train))
print(len(df_test))
```

```python
training_set = df_train['Close'].values
training_set
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
training_set_scaled = scaler.fit_transform(training_set.reshape(-1,1))
training_set_scaled
```

```python
intervalo_entre_dias = 60
x_train = []
y_train = []

for i in range(intervalo_entre_dias,len(training_set)):
  x_train.append(training_set_scaled[i-intervalo_entre_dias:i,0])
  y_train.append(training_set_scaled[i,0])
import numpy as np
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

# Treino

```python
batch_size = len(x_train)
time_steps = intervalo_entre_dias
features = 1

x_train = np.reshape(x_train,(batch_size,time_steps, features))
```

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

rnn = Sequential()
```

```python
rnn.add(LSTM(units=50,return_sequences=True,input_shape=(time_steps, features)))
rnn.add(Dropout(0.5))

rnn.add(LSTM(units=50,return_sequences=True))
rnn.add(Dropout(0.5))

rnn.add(LSTM(units=50,return_sequences=True))
rnn.add(Dropout(0.5))

rnn.add(LSTM(units=50))
rnn.add(Dropout(0.5))

rnn.add(Dense(units=1))
```

```python
from tensorflow.keras.utils import plot_model
plot_model(rnn, to_file='modelo.png', show_shapes=True, show_layer_names=True, expand_nested=True)
```

```python
rnn.compile(optimizer='adam',loss='mean_squared_error', metrics=['mean_absolute_error'])
rnn.fit(x_train,y_train,epochs=100,batch_size=32)
```

# Teste

```python
fechamento_real = df_test['Close'].values

dataset_total = pd.concat((df_train['Close'],df_test['Close']),axis=0)
inputs = dataset_total[len(dataset_total)-len(df_test)-intervalo_entre_dias:].values
inputs = inputs.reshape(-1,1)
inputs = scaler.transform(inputs)
inputs = scaler.transform(inputs)
```

```python
x_test = []
for i in range(intervalo_entre_dias,len(inputs)):
  x_test.append(inputs[i-intervalo_entre_dias:i,0])

x_test = np.array(x_test)
x_test = np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
```

```python
print(f'previsao oficial: {rnn.predict(x_test)}')
print(f'previsao real: {scaler.inverse_transform(rnn.predict(x_test))}')

variacao_final = scaler.inverse_transform(rnn.predict(x_test))
print(variacao_final)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=list(range(len(fechamento_real))), y=fechamento_real, mode='lines', name=f'Dados reais das acoes nvd',line=dict(color='red')))

fig.add_trace(go.Scatter(x=list(range(len(variacao_final))), y=variacao_final, mode='lines', name=f'Dados previsto das acoes nvda',line=dict(color='blue')))

fig.update_layout(title='Previsao de preços de fechamento',xaxis_title='Tempo',yaxis_title='Preço de fechamento', legend_title = "Legenda")
fig.show()
```
